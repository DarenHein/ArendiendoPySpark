```python
from pyspark.sql import SparkSession 
from pyspark.sql.functions import * 
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, FloatType
```

```python
'''
StructType y StructField

es como el contrato de como viene nuestro csv o parquete los nombres de las columnas 
su tipo de dato y si acepta o no nulos 

'''
#todo ejemplo de structype y structuFiedl 
#todo Definimos el Schema 
'''
 |-- id: string (nullable = true)
 |-- Index: string (nullable = true)
 |-- Name: string (nullable = true)
 |-- Form: string (nullable = true)
 |-- Type1: string (nullable = true)
 |-- Type2: string (nullable = true)
 |-- HP: string (nullable = true)
 |-- Attack: string (nullable = true)
 |-- Defense: string (nullable = true)
 |-- SpAtk: string (nullable = true)
 |-- SpDef: string (nullable = true)
 |-- Speed: string (nullable = true)
'''
schema_pokemon = StructType([
    StructField("id", IntegerType() , False ),
    StructField("index" , StringType() , False ), 
    StructField("Name" , StringType(), False ), 
    StructField("Form", StringType() , False ),
    StructField("Type1", StringType() , False ),
    StructField("Type2", StringType() , True ),
    StructField("HP", StringType() , False ),
    StructField("Attack", StringType() , False ),
    StructField("Defense", StringType() , False ),
    StructField("SpAtk", StringType() , False ),
    StructField("SpDef", StringType() , False ),
    StructField("Speed", StringType() , False ),
])


spark = SparkSession.builder \
    .appName("MiPrimeraVezConSpark") \
    .getOrCreate()
```

LECTURA DE INSUMOS

```python
# df = spark.read.option("header", "true").csv("pokemon.csv")
#todo aplicando nuestro schema al csv de pokemon 
df = spark.read \
.format("csv") \
.option("header","true") \
.schema(schema_pokemon) \
.load("pokemon.csv")

df_parquet = spark.read.parquet("pokemon.parquet")
```

```python
df.printSchema()
```

FUNCIONES 

```python
def FiltradoPorTipo(df):
    df.filter(col("Type1") == "Fire")
    return df 
```

ORQUESTADOR 

```python
dfFiltrado = FiltradoPorTipo(df)
```

TRANSFOMRACION A PARQUET 

```python
'''
Tipos de mode para la creacion de un parquet 
overwrite -> sobreescribe lo que hay 
append    -> agrega al archivo 
ignore    -> ingnora si ya existe 
'''
def parquetCreate(df , mode , name):
    df.write.mode(mode).parquet(name)

parquetCreate(df,"overwrite","pokemon.parquet")
```
